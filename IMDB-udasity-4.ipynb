```python
data = pd.read_csv('imdb.csv') #read file
data.head()
```

```python
p = data[['runtime', 'original_title']] #select runtime &original_title
p.sort_values(by=["runtime"], ascending=False)  #Which original_title are the most runtime of all time?
```

```python
r = data[['revenue', 'original_title']] #select revenue &original_title
r.sort_values(by=["revenue"], ascending=False)  #Which original_title are the most revenue of all time?
```

```python
t = data[['budget', 'original_title']] #select budget &original_title
t.sort_values(by=["budget"], ascending=False)  #Which original_title are  top budget of all time?
```

```python
z = data[['vote_count', 'original_title']] #select vote_count &original_title
z.sort_values(by=["vote_count"], ascending=False)  #Which original_title are the most vote_count of all time?
```

```python
y = data[['popularity', 'cast']] #select popularity & cast
y.sort_values(by=["popularity"], ascending=False)  #Which cast are the most popular of all time?

```
